## Let's compare stemmers!
First load in the libraries and set up the stemmers.

```python
import nltk
from nltk.corpus import stopwords
english_stopwords = stopwords.words('english')

from nltk.corpus import gutenberg
import re
import string
pstemmer = nltk.PorterStemmer()
lstemmer = nltk.LancasterStemmer()
wnlemmatizer = nltk.WordNetLemmatizer()
```

We'll also need a bit of tokenization and normalization here.  
**Add some code here to remove punctuation and stopwords.**

```python
def tokenize_text(text):    
    pattern = r'''(?x)
    ([A-Z]\.)+
    |\w+([-']\w+)*
    |\$?\d+(\.\d+)?%?
    |\.\.\.
    |[.,?;]+
    '''
    tokens = nltk.regexp_tokenize(text,pattern)
    # add code to remove punctuation and stopwords
    return tokens
```

Let's tokenize Shakespeare's Julius Ceaser; make sure it worked.

```python
tokens = tokenize_text(gutenberg.raw('shakespeare-caesar.txt'))
print (len(tokens))
print(tokens[0:10])
```

Here is one example of running the stemmer; ** you can fill in the other two.**

```python
pstemmed_tokens = [pstemmer.stem(word.lower()) for word in tokens if word.isalpha() and word not in english_stopwords]
lstemmed_tokens = [lstemmer.stem(word.lower()) for word in tokens if word.isalpha() and word not in english_stopwords]
wnlemma_tokens = [wnlemmatizer.lemmatize(word.lower()) for word in tokens if word.isalpha() and word not in english_stopwords]
```

**Now we want make 3 FreqDists to count up how common each word is now that we've done the stemming.**

```python
fdist1 = nltk.FreqDist(pstemmed_tokens)
fdist1.most_common()
fdist2 = nltk.FreqDist(lstemmed_tokens)
fdist2.most_common()
fdist3 = nltk.FreqDist(wnlemma_tokens)
fdist3.most_common()
```

Now we want to compare the three outputs.  **What is a good function that quickly lets you do side-by-side comparisons of vertical lists?**

```python
zipped = zip(fdist1.most_common(20), fdist2.most_common(20), fdist3.most_common(20))
list(zipped)
```
